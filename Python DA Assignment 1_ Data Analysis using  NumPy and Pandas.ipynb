NUMPY ARRAY OPERATIONS:

```python
import numpy as np
```

```python
#1. Creation of a 1D NumPy Array:

temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])

print(temperatures_w1)
```

```python
#2. Inspection and Properties:

print("Shape:", temperatures_w1.shape)
print("Datatype:", temperatures_w1.dtype)
print("Number of Elements:", temperatures_w1.size)
```

```python
#3. Array Operations:

#Converting from Celsius to Fahrenheit:
temperatures_in_fahrenheit = (temperatures_w1 * 9/5) + 32
print("Converted to Fahrenheit:", temperatures_in_fahrenheit)

#Finding the maximum, minimum, and mean temperatures for the week:
max_temp = np.max(temperatures_w1)
print("Maximum Temperature:", max_temp)
min_temp = np.min(temperatures_w1)
print("Minimun Temperature:", min_temp)
mean_temp = np.mean(temperatures_w1)
print("Mean Temperature:", mean_temp)
```

```python
#4. Array Slicing and Indexing:

#First three days:
First_three_Days = temperatures_w1[0:3]
print("First three days:", First_three_Days)
#Temperatures for the weekend:
weekend_temperatures = temperatures_w1[-2:]
print("Temperatures for the weekend:", weekend_temperatures)
#Temperatures for the middle three days of the week:
middle_three_days = temperatures_w1[2:5]
print("Temperatures for the middle three days of the Week:", middle_three_days)
```

```python
#5. Creation of a 2D Array:

temperatures ={
               "Week 1": np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9]),
               "Week 2": np.array([19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1])
}
print("Week 1:", temperatures["Week 1"])
print("Week 2:", temperatures["Week 2"])

temperatures_2d = np.array(list(temperatures.values()))
```

```python
#6. Inspect and Slice the 2D Array:

#Inspecting and printing the shape, data type, and total number of elements:
print("Shape:", temperatures_2d.shape)
print("Data Type:", temperatures_2d.dtype)
print("Total number of Elements:", temperatures_2d.size)
```

```python
#Extract the temperatures for each week and the weekends for both weeks:
week1 = temperatures[0]
week2 = temperatures[1]
print("Week 1:", temperatures[0])
print("Week 2:", temperatures[1])

print("Week 1 Weekend:", temperatures[0, -2:])
print("Week 2 Weekend:", temperatures[1, -2:])
```

PANDAS SERIES:

```python
import pandas as pd
```

```python
#1. Creating a Pandas Series:
marks = pd.Series([95, 92, 89, 85, 80],index = ["Rank1", "Rank2", "Rank3","Rank4", "Rank5"])
print(marks)
```

```python
#2. Indexing and Slicing:

#Access the mark of the 1st rank student:
first_rank_student = marks.iloc[0]
print("Mark of the 1st rank student:", first_rank_student)

#Retrieve the marks of the top 3 ranks:
print("Marks of the Top 3 Ranks:")
print(marks.loc[['Rank1', 'Rank2', 'Rank3']])

#Retrieve the mark of the 3rd rank student:
third_rank_mark = marks.iloc[2]
print("Mark of the 3rd rank student:", third_rank_mark)

#Retrieve the ranks where the marks are greater than 90:
print("Marks greater than 90:")
print(marks[marks > 90])
```

```python
#3. Manipulating Series:

#Modify the mark of the student with the 1st rank to 100:
marks["Rank1"] = 100
print(marks)
```

```python
#Remove the entry corresponding to the last rank from the Series:
marks = marks.drop("Rank5")
print(marks)
```

```python
#Compute the CGPA by dividing each mark in the Series by 10:
cgpa = marks / 10
print("CGPA:")
print(cgpa)
```

PANDAS DATA FRAME:



```python
#1. Creating Pandas DataFrame:

transactions = pd.DataFrame({
    'Transaction ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product Category': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    'Region': ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South', 'North'],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]})
print(transactions)
```

```python
#2. Data Exploration:

#Display the transactions DataFrame and its basic information:
print("Head:", transactions.head(1))

print("Tail:", transactions.tail(1))

print("Shape:", transactions.shape)

print("Column Names:", transactions.columns)

print("Data Types:", transactions.dtypes)
```

```python
#Display only the 'ProductCategory' and 'Amount' columns:
print(transactions[['Product Category', 'Amount']])
```

```python
print(transactions.loc[:, ["Product Category", "Region", "Amount"]])
```

```python
#Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200:
print(transactions[(transactions['Region']=='North') & (transactions['Amount']>200)])
```

```python
#Find the value counts for the 'ProductCategory' column:
print(transactions['Product Category'].value_counts())
```

```python
#Find the unique values in the 'Region' column:
print(transactions['Region'].unique())
```

```python
#Group by 'Region' and find the mean amount for each region:
print(transactions.groupby('Region')['Amount'].mean())
```

```python
#3. Manipulating the DataFrame:

#Modify the 'Amount' for the transaction with TransactionID 102 to 165:
transactions.loc[transactions['Transaction ID']==102, 'Amount']=165
print(transactions)
```

```python
#Add a new column 'Discount' by calculating 10% of the 'Amount':
transactions['Discount'] = 0.1 * transactions['Amount']
print(transactions)
```

```python
#Remove the row with TransactionID 109:
transactions = transactions[transactions['Transaction ID'] != 109]
print(transactions)
```

```python
#Delete the 'Discount' column from the DataFrame:
transactions = transactions.drop(columns=['Discount'])
print(transactions)
```
